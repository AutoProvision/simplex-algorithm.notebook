##**AMPLITUDE**

**Amplitude**: Permite comparar a variabilidade entre diferentes características das flores. Por exemplo, se uma característica tem uma amplitude muito maior que outra, isso indica que os valores dessa característica variam mais amplamente.

```python
# Carregar o dataset Iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Calcular amplitude (range) para cada característica
range_values = df.max() - df.min()

# Plotar gráfico de amplitude
plt.figure(figsize=(7, 6))
plt.bar(range_values.index, range_values.values, color='skyblue')
plt.title('Amplitude (Range) das Características')
plt.ylabel('Amplitude (cm)')
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

##**ASSIMETRIA**

**Assimetria**: Mede a simetria de um conjunto de dados em relação a média. Podendo ser:


*   **Assimetria positiva:** Cauda mais longa a direita.
*   **Assimetria negativa:** Cauda mais longa a esquerda.

Ex.: Um valor negativo de assimetria sugere que a maioria dos dados está concentrada em valores mais altos, com uma cauda estendida de valores mais baixos. Esse insight é útil para entender a distribuição dos dados e fazer escolhas informadas sobre como resumir ou modelar os dados.



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew
from sklearn.datasets import load_iris

# Carregar o dataset Iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Calcular assimetria (skewness) para cada característica
skewness_values = df.apply(skew)

# Plotar gráfico de assimetria
plt.figure(figsize=(7, 6))
plt.bar(skewness_values.index, skewness_values.values, color='lightcoral')
plt.title('Assimetria (Skewness) das Características')
plt.ylabel('Assimetria')
for index, value in enumerate(skewness_values):
    plt.text(index, value, f'{value:.2f}', ha='center', va='bottom')
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

# LEANDRO

```python
import pandas as pd
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kurtosis
```

```python
def calcular_coeficiente_variacao(dados):

    media = np.mean(dados)
    desvio_padrao =  np.std(dados)
    coeficiente_variacao = (desvio_padrao / media) * 100

    return coeficiente_variacao

iris = datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
df['species'] = df['target'].map(lambda x: iris.target_names[x])
```

##**COEFICIENTE DE VARIAÇÃO**

```python
grupo_esp = df.groupby('species')

resultados = {}

for especie in grupo_esp.groups:

    grupo = grupo_esp.get_group(especie)

    resultados[especie] = {
        'coeficiente_petal_length': calcular_coeficiente_variacao(grupo['petal length (cm)']),
        'coeficiente_petal_width': calcular_coeficiente_variacao(grupo['petal width (cm)']),
        'coeficiente_sepal_length': calcular_coeficiente_variacao(grupo['sepal length (cm)']),
        'coeficiente_sepal_width': calcular_coeficiente_variacao(grupo['sepal width (cm)'])
    }

labels = []
valores = []

for especie, coeficientes in resultados.items():

    for caracteristica, valor in coeficientes.items():

        labels.append(f'{especie} - {caracteristica}')
        valores.append(valor)

plt.figure(figsize=(10, 6))
plt.barh(labels, valores, color='skyblue')

for index, value in enumerate(valores):
    plt.text(value, index, f'{value:.2f}%', va='center')

plt.xlabel('Coeficiente de Variação (%)')
plt.ylabel('Características por Espécie')
plt.title('Coeficiente de Variação por Característica e Espécie')
plt.show()
```

## **CURTOSE/KURTOSIS**

```python
curtose = kurtosis(df['sepal width (cm)'])

plt.hist(df['petal length (cm)'], bins=5, color='lightcoral')
plt.title(f'Curtose = {curtose:.2f}')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```
