```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 35, 10)

plt.figure(figsize=(10, 6))
plt.ioff();
```

Variáveis:

$$x = Pandas$$

$$y = PySpark$$

Restrição 1:

$$x + y = 30$$

```python
y1 = 30 - x

plt.plot(x, y1, label=r'$x + y \leq 30$');
```

Restrição 2:

$$120⋅x+125⋅y ≤ 4,000$$

```python
y2 = (4000 - 120*x) / 125

plt.plot(x, y2, label=r'$120 \cdot x + 125 \cdot y \leq 4,000$');
```

Restrição 3:

$$5 ≤ x$$

$$5 ≤ y$$

```python
plt.axvline(5, color='gray', linestyle='--', label=r'$5 \leq x$')
plt.axhline(5, color='gray', linestyle='-.', label=r'$5 \leq y$')
```

Traçando Conjunto Solução

```python
def min_max(var, min_val, max_val):
	return np.minimum(max_val, np.maximum(min_val, var))

x_sides = np.maximum(x, 5)
y_lower = min_max(y1, 5, 30)
y_upper = min_max(y2, 5, 30)

plt.fill_between(x_sides, y_lower, y_upper, where=(y_upper >= y_lower), color='gray', alpha=0.3);
```

```python
plt.xlim(0, 35)
plt.ylim(0, 35)

plt.xlabel(r'$x$ (Chamadas pelo Pandas)')
plt.ylabel(r'$y$ (Chamadas pelo PySpark)')

plt.legend()

plt.title('Otimização de Chamadas da camada Trusted')
plt.grid(True)
plt.show()
```
