Este problema com apenas duas variáveis pode ser resolvido graficamente. Traça-se um gráfico com os seus eixos sendo as duas variáveis ou mais variáveis. A partir daí, traçam-se as retas referentes às restrições do problema e delimita-se a região viável, denominada conjunto solução.

Encontrada a região viável, deve-se traçar uma reta com a inclinação da função objetivo. São então traçadas diversas paralelas a ela no sentido de Z crescente (maximização da função). Encontre a figura correta para o problema de PO a seguir, com a execução do código em python.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linprog
# Utilizamos a lineprog para minimizar uma função objetivo linear sujeita a restrições lineares de igualdade e desigualdade
from matplotlib.patches import Polygon
%matplotlib inline
```

```python
# Construindo as linhas

# x > 0

x = np.linspace(0, 30, 2000) # retorna números com espaçamento uniforme em um intervalo especificado

# y >= 3

y1 = (x*0) + 3

# 3y <= 36 - x

y2 = (36-x)/3.0

# 6y >= 2x - 8

y3 = (2*x-8)/6.0

# y <= 2x - 12

y4 = 2 * x -12
```

```python
# plotando o gráfico

plt.plot(x, y1, label=r'$y\geq3$')

plt.plot(x, y2, label=r'$3y\leq36-x$')

plt.plot(x, y3, label=r'$6y\geq 2x - 8$')

plt.plot(x, y4, label=r'$y\leq 2x-12$')

plt.xlim((0, 30))

plt.ylim((0, 25))

plt.xlabel(r'$x$')

plt.ylabel(r'$y$')

# Preenchendo a região viável

y5 = np.minimum(y2, y4)

y6 = np.maximum(y1, y3)

plt.fill_between(x, y5, y6, where=y5>y6, color='grey', alpha=0.5)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
