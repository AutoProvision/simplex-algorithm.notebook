O método simplex pode ser utilizado quando existem apenas poucas variáveis no problema em geral até 3, desta maneira podemos resolver o conjunto de inequações pelo método gráfico ou pela substituição e isolamento das variáveis.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 10000)
```

```python
y1 = 11.0 + x
y2 = 27.0 - x
y3 = (90 - 2 * x) / 5.0
```

```python
plt.plot(x, y1, label=r'$-x+y\leq11$')
plt.plot(x, y2, label=r'$x+y\leq27$')
plt.plot(x, y3, label=r'$2x+5y\leq90$')

plt.xlabel(r'$x$')
plt.ylabel(r'$y$')

y5 = np.minimum(y1, y2)

plt.fill_between(x, y3, y5, where=y3<y5, color='gray', alpha=0.5)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```
